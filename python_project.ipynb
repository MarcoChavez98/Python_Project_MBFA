```python
#Chargement des données

import pandas as pd

df = pd.read_excel(r"C:\Users\ccraq\OneDrive\Documents\PYTHON\projet_python\dataset.xlsx")

data= pd.DataFrame(df, columns=['id', 'imdb_id', 'popularity','budget','revenue','original_title','cast','homepage','director','tagline','keywords','overview','runtime','genres','production','release_date','vote_count','vote_average','release_year','budget_adj','revenue_adj'])

```

```python
#On enlève les colonnes que l'on ne souhaite pas exploiter

data = data.drop(columns=['imdb_id','popularity','homepage','release_date','budget_adj','revenue_adj'])
data
```

```python
#Enlever les films en doublon 
data["original_title"].drop_duplicates
```

```python
#On créée un dataframe qui sera utilisé pour les content based recommendations

movies_data=data.loc[:,["original_title","cast","genres","tagline","keywords","overview"]]

movies_data.head()
movies_data.shape
```

```python
#On regarde combien il y a de valeurs nulles

movies_data.isna().sum()
```

```python
#On enlèves les valeurs nulles

movies_data.dropna(inplace=True)
```

```python
#On vérifie bien qu'il n'y en a plus
movies_data.isna().sum()
```

```python
#Si on regarde au dessus, les colonnes sont devenues des integer ce qui est problématique donc on les transforme en strings

movies_data= movies_data.astype(str)

movies_data.head()
```

```python
#On remplace les "|" par des espaces poyur les colonnes concernées

movies_data["cast"] = movies_data["cast"].apply(lambda x: x.replace("|"," "))
movies_data["genres"] = movies_data["genres"].apply(lambda x: x.replace("|"," "))
movies_data["keywords"] = movies_data["keywords"].apply(lambda x: x.replace("|"," "))
movies_data["genres"] = movies_data["genres"].apply(lambda x: x.replace("|"," "))

movies_data.head()

```

```python
#Ici on combine les acteurs+genres pour pouvoir regrouper les films qui se ressemblent entre eux, en plus j'ai aussi joint les descriptifs pour comparer les mots utilisés

movies_data["combined"]=movies_data["genres"] + " " + movies_data["cast"]
movies_data.head()

movies_data["combined2"]=movies_data["tagline"] + " " + movies_data["keywords"] + " " + movies_data["overview"]
movies_data.head()
```

```python
#On change le nom du titre pour plus de clareté et on les mets en minuscules

movies_data.rename(columns={"original_title":"title"},inplace=True)
movies_data.head()


movies_data["title"] = movies_data["title"].str.lower()
movies_data.head()

movies_data.info()
```

```python
#On est obligés de faire ça pour éviter les beugs dus au fait d'avoir enlever les valeurs NaN

movies_data.reset_index(drop=True,inplace=True)
```

```python
#datacamp

from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate the vectorizer object and transform the plot column
vectorizer = TfidfVectorizer(max_df=0.7, min_df=10)
vectorized_data = vectorizer.fit_transform(movies_data['combined']) 

# Create Dataframe from TF-IDFarray
tfidf_df = pd.DataFrame(vectorized_data.toarray(), columns=vectorizer.get_feature_names())

# Assign the movie titles to the index and inspect
tfidf_df.index = movies_data['title']
print(tfidf_df.head())
```

```python
#datacamp

# Import cosine_similarity measure
from sklearn.metrics.pairwise import cosine_similarity

# Create the array of cosine similarity values
cosine_similarity_array = cosine_similarity(tfidf_df)

# Wrap the array in a pandas DataFrame
cosine_similarity_df = pd.DataFrame(cosine_similarity_array, index=tfidf_df.index, columns=tfidf_df.index)

# Print the top 5 rows of the DataFrame
print(cosine_similarity_df.head())
```
